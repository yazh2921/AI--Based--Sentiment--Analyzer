```python
from textblob import TextBlob

while True:
    text = input("Enter review (or type exit): ")

    if text.lower() == "exit":
        break

    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        print("Sentiment: Positive 😊")
    elif polarity < 0:
        print("Sentiment: Negative 😔")
    else:
        print("Sentiment: Neutral 😐")

    print("------------------")
```
